# **SVR (Support Vector Regression)**

**Feature Scaling is a must and is done before spiltting the data**

```python

X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
X[:,0]=encoder.fit_transform(X[:,0])
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler for height feature
sc_height_X = StandardScaler()
X[:, 1] = sc_height_X.fit_transform(X[:, 1].reshape(-1, 1)).reshape(-1)

# Initialize StandardScaler for weight feature
sc_weight_X = StandardScaler()
X[:, 2] = sc_weight_X.fit_transform(X[:, 2].reshape(-1, 1)).reshape(-1)

# Initialize StandardScaler for target (BMI Index)
sc_index_y = StandardScaler()
y = sc_index_y.fit_transform(y.reshape(-1, 1)).reshape(-1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf')
svr=regressor.fit(X_train,y_train)
```

```python
svr_pred=np.round(sc_index_y.inverse_transform(svr.predict(X_test).reshape(-1,1)))
```

```python
#Comparision
print(np.concatenate((svr_pred.reshape(len(svr_pred), 1), sc_index_y.inverse_transform(y_test.reshape(-1, 1))), axis=1))
```

```python
import matplotlib.pyplot as plt
# Generate indices for x-axis
indices = np.arange(len(y_test))

# Plot actual values
plt.plot(indices, sc_index_y.inverse_transform(y_test.reshape(-1,1)), label='Actual', marker='o')

# Plot predicted values
plt.plot(indices, svr_pred, label='Predicted', marker='^')

plt.xlabel('Sample Index')
plt.ylabel('BMI Index')
plt.title('Actual vs Predicted BMI Index (SVR)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(sc_index_y.inverse_transform([y_test]).reshape(-1),svr_pred.reshape(-1))
```

```python
#Accuray
svr.score(X_test,y_test)
```

# **Random Forest regression**

```python
from sklearn.model_selection import train_test_split
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
X[:,0]=encoder.fit_transform(X[:,0])

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)

from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=10,random_state=0)
rf_reg=rf.fit(X_train,y_train)

rf_pred=np.round(rf_reg.predict(X_test))

#Comparsion btw sample values and predicted values
print(np.concatenate((rf_pred.reshape(len(rf_pred),1),y_test.reshape(len(y_test),1)),1))
```

**Visualization of Random Forest Regressor**

```python
# Generate indices for x-axis
indices = np.arange(len(y_test))

# Plot actual values
plt.plot(indices, y_test, label='Actual', marker='o')

# Plot predicted values
plt.plot(indices, rf_pred, label='Predicted', marker='^')

plt.xlabel('Sample Index')
plt.ylabel('BMI Index')
plt.title('Actual vs Predicted BMI Index (Multiplr Regression)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,rf_pred)
```

```python
#Accuracy
rf_reg.score(X_test,y_test)
```
